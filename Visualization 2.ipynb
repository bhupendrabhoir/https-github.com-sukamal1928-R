# Visualization 2

## Two variable plot

```r
df <- read.csv("house.csv")
```

```r
head(df)
```

```r
library(ggplot2)
```

### 2d Bin plot

```r
pl <- ggplot(df,aes(x=sqft_living,y=price))
pl + geom_bin2d()
```

```r
pl <- ggplot(df,aes(x=sqft_living,y=price))
pl <- pl + geom_bin2d()
pl + scale_fill_gradient(high='orange',low='lightgreen')
```

### 2d Density plot

```r
pl <- ggplot(df,aes(x=sqft_living,y=price))
pl + geom_density2d()
```

### 2d hex plot

```r
pl <- ggplot(df,aes(x=sqft_living,y=price))
pl + geom_hex()
```

```r
pl <- ggplot(df,aes(x=sqft_living,y=price))
pl <- pl + geom_hex()
pl + scale_fill_gradient(high='blue',low='lightblue')
```

## Coordinates

```r
pl <- ggplot(df,aes(x=bedrooms,y=bathrooms))
pl <- pl + geom_point()
print(pl)
```

```r
pl <- ggplot(df,aes(x=bedrooms,y=bathrooms))
pl <- pl + geom_point()
pl <- pl + coord_cartesian(xlim=c(1,15),ylim=c(1,4))
print(pl)
```

```r
pl <- ggplot(df,aes(x=bedrooms,y=bathrooms))
pl <- pl + geom_point()
pl <- pl + coord_fixed(ratio=3/2)
print(pl)
```

## Faceting

```r
df <- read.csv("cars.csv")
```

```r
head(df)
```

```r
p <- ggplot(df, aes(engine.size, price)) + geom_point()

p + facet_grid(. ~ body.style)
```

```r
p <- ggplot(df, aes(engine.size, price)) + geom_point()

p + facet_grid(body.style ~ .)
```

```r
p <- ggplot(df, aes(engine.size, price)) + geom_point()

p + facet_grid(drive.wheels ~ body.style)
```

## Theme

```r
p <- ggplot(df, aes(engine.size, price)) + geom_point()

p <- p + facet_grid(. ~ body.style)

p <- p + theme_bw()

print(p)
```

```r
p <- ggplot(df, aes(engine.size, price)) + geom_point()

p <- p + facet_grid(. ~ body.style)

p <- p + theme_classic()

print(p)
```

```r
p <- ggplot(df, aes(engine.size, price)) + geom_point()

p <- p + facet_grid(. ~ body.style)

p <- p + theme_dark()

print(p)
```
