# Visualization 1

```r
df = read.csv("cars.csv")
head(df)
```

```r
summary(df)
```

```r
# Basic plot

hist(df[["price"]])
```

## ggplot 

```r
library(ggplot2)
```

### 1) Histogram

```r
# data and aesthetics
pl <- ggplot(df,aes(x=price))

#geometry

pl <- pl + geom_histogram()

print(pl)
```

```r
# data and aesthetics
pl <- ggplot(df,aes(x=price))

#geometry

pl <- pl + geom_histogram(fill="grey",color="black",binwidth=500)

print(pl)
```

```r
# data and aesthetics
pl <- ggplot(df,aes(x=width))

#geometry

pl <- pl + geom_histogram(fill="skyblue",color="skyblue",binwidth=0.4) + xlab("width") + ylab("frequency") + ggtitle("width vs frequency")
print(pl)
```

```r
# data and aesthetics
pl <- ggplot(df,aes(x=width))

#geometry

pl <- pl + geom_histogram(aes(fill=..count..),color="skyblue",binwidth=0.4) + xlab("width") + ylab("frequency") + ggtitle("width vs frequency")

print(pl)
```

```r
# data and aesthetics
pl <- ggplot(df,aes(x=width))

#geometry

pl <- pl + geom_histogram(aes(fill=..count..),color="skyblue",binwidth=0.3) + xlab("width") + ylab("frequency") + ggtitle("price vs frequency")

pl = pl + scale_fill_gradient('count',high='lightblue',low='lightgreen')

print(pl)
```

### 2) Scatter plot

```r
pl <- ggplot(df,aes(x=engine.size,y=price))

pl <- pl + geom_point() 
print(pl)
```

```r
pl <- ggplot(df,aes(x=engine.size,y=price))

pl <- pl + geom_point(color="blue") + xlab("engine size") + ylab("price") + ggtitle("engine size vs price")

print(pl)
```

```r
pl <- ggplot(df,aes(x=engine.size,y=price))

pl <- pl + geom_point(aes(color=width)) + xlab("engine size") + ylab("price") + ggtitle("engine size vs price")

print(pl)
```

```r
pl <- ggplot(df,aes(x=engine.size,y=price))

pl <- pl + geom_point(aes(color=width,size=factor(drive.wheels)),alpha=0.4) + xlab("engine size") + ylab("price") + ggtitle("engine size vs price")

pl <- pl + scale_colour_gradient(high="red",low ="blue")

print(pl)
```

```r
pl <- ggplot(df,aes(x=engine.size,y=price))

pl <- pl + geom_point(aes(color=width,shape=factor(drive.wheels)),size=4,alpha=0.4) + xlab("engine size") + ylab("price") + ggtitle("engine size vs price")

print(pl)
```

### 3) Bar plot

```r
g <- ggplot(df, aes(body.style))

g + geom_bar()
```

```r
g <- ggplot(df, aes(body.style))

g + geom_bar(aes(fill = drive.wheels))
```

```r
g <- ggplot(df, aes(body.style))

g + geom_bar(aes(fill = drive.wheels),position="dodge") + xlab("body style") + ylab("count") + ggtitle("body style vs count")
```

```r
g <- ggplot(df, aes(body.style))

g + geom_bar(aes(fill = drive.wheels),position="fill") + xlab("body style") + ylab("count") + ggtitle("body style vs count")
```

### 4) Box plot

```r
pl <- ggplot(df, aes(engine.type,price))

pl + geom_boxplot()
```

```r
pl <- ggplot(df, aes(engine.type,price))

pl + geom_boxplot() + coord_flip()
```

```r
pl <- ggplot(df, aes(engine.type,price))

pl + geom_boxplot(color="grey",fill="lightblue")
```

```r
pl <- ggplot(df, aes(engine.type,price))

pl + geom_boxplot(aes(fill=engine.type))
```
