# Kmeans clustering

```r
library(ggplot2)
```

```r
df <- read.csv("mall.csv")
```

```r
head(df)
```

```r
ggplot(df, aes(INCOME,SPEND)) + geom_point()
```

```r
set.seed(101)
```

```r
df.clusters <- kmeans(df, 6, nstart = 20)
```

```r
df.clusters
```

```r
df.clusters$cluster
```

```r
library(cluster)
```

```r
clusplot(df, df.clusters$cluster, color=TRUE, shade=TRUE, labels=0,lines=0, )
```
